### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# 1. FIGURE: 
#    Think of the figure as a blank sheet of paper or a canvas. 
#    It's the overall space where you want to start drawing or plotting.
 
# 2. AXES: 
#    On that piece of paper, the axes are like specific areas where you decide 
#    to place your drawings or plots. It's a designated section of the paper 
#    for your artwork or data.

# 3. AXIS: 
#    Now, within those designated sections, you have reference lines to help guide your drawing or plot. 
#    These lines help position data points. 
#    Typically, the x-axis is the line that runs left-to-right (like the timeline on a graph), 
#    and the y-axis is the one that runs up-and-down (like how tall something is on a graph).

# 4. ARTIST
#    In Matplotlib, "artist" is a fundamental elements used to create visual 
#    components on a canvas, such as lines, shapes, and text. Everything seen 
#    in a Matplotlib figure is an artist,including the figure itself, axes, labels, 
#    and plots.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4,4,10)   # We create a numpy array with 10 random values between -4,4
y = x**2
                           # For both approach we have the same beginning
    
# We want boht approach in one row side by side
# Object oriented approach

fig,ax = plt.subplots(1, 2, figsize=(10, 5), constrained_layout=True)
ax[0].plot(x, y, label='y^2', color='green')
ax[0].set_title(' y(x) = x^2 ', color='blue')
ax[0].set_xlabel('x', color='blue')
ax[0].set_ylabel('y', color='blue')
ax[0].legend('Object')

# Second plot (using pyplot style)
plt.plot(x, y, color='orange')
plt.xlabel('X', color='green')
plt.ylabel('Y', color='green')
plt.title('y(x) = x^2', color='green')
plt.legend('Pyplot')

plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)          # Not needed or used
y = x**2                             # Not needed or used

# Data for bar plot
fruit_data = {'grapes': 20, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
# Create a figure with 1 row and 2 columns for subplots
fig, ax = plt.subplots(1, 2, figsize=(15, 5), layout ='constrained')

# Scatter Plot
ax[0].scatter(random_data_x,random_data_y, color = 'green')
ax[0].set_xlabel('x', color='blue')
ax[0].set_ylabel('y', color='blue')
ax[0].set_title('VALUE OF FRUITS IN ICA', color ='black')
ax[0].legend(['SCATTER'])

# 2 - Barplot
ax[1].bar(names,values,color='blue')
ax[1].set_title('VALUE OF FRUITS IN ICA')
ax[1].set_xlabel('FRUITS')
ax[1].set_ylabel('KR')
ax[1].legend(['BAR PLOT'])
plt.show()
```
